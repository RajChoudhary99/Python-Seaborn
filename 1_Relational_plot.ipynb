Relational Plot

1) to see the statistical relation between 2 or more variables.
2) Bivariate Analysis

Plots under this section -

scatterplot
lineplot

figure level -> relplot
axes level -> scatterplot, lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
temp_df1 = gap[gap['country'] == 'India']
temp_df1
```

```python
# axes level function
sns.lineplot(data=temp_df1, x='year', y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=temp_df1, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style, multipole lines
temp_df2 = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df2
```

```python
sns.relplot(kind='line', data=temp_df2, x='year', y='lifeExp', hue='country', style='continent') #as you can see here , the legend is outside the graph as figure level functions have the access of whole chart so legend can be placed anywhere
```

```python
sns.lineplot(data=temp_df2, x='year', y='lifeExp', hue='country')# here the legend is inside the graph as axes level functions have access to only the axes so legend is placed inside the axes
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='smoker')
```
