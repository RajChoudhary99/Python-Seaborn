Categorical Plots

1) Categorical Scatter Plot- used when we want to compare a categorical col with a numerical col, unlike normal scatter plot which only compares two numerical cols. it is bivariate

Stripplot
Swarmplot

2) Categorical Distribution Plots- comapare categorical with numerical, it's both uni and bivariate

Boxplot
Violinplot

3) Categorical Estimate Plot -> for central tendency -

Barplot
Pointplot
Countplot

4) Figure level function -> catplot

```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

1) Categorical Scatter plots

```python
sns.scatterplot(data=tips, x='total_bill',y='tip') #just for comparison
```

```python
# strip plot
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=False,hue='sex')
```

```python
#swarm plot
#axes level function
sns.swarmplot(data=tips, x='day',y='total_bill')
```

```python
#figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```

```python
# hue
sns.swarmplot(data=tips, x='day',y='total_bill',hue='sex')
```

2) Categorical Distribution Plots

Boxplot - 

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# Using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

Violinplot = (Boxplot + KDEplot)

boxplot is the line in middle and kde is the violin shape

```python
# violinplot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue

sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

3) Categorical Estimate Plot

```python
#barplot
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
#pointplot (used for connecting points of barplot)
sns.pointplot(data=tips,  x='sex', y='total_bill')
```

```python
# countplot(A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable)
sns.countplot(data=tips,x='sex')
```

```python
# faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',col='smoker',kind='box',row='time')
```
